```python
import csv
import pandas as pd
import numpy as np
from numpy import *
%matplotlib inline
import matplotlib.pyplot as plt
```

**Импортируем данные о пользователях и просмотренных ими фильмах:**

```python
user_item = pd.read_csv('train_likes.csv')
user_item = pd.DataFrame(user_item, columns=['user_id', 'item_id'])
user_item.head()
```

**Ограничим количество пользователей в выборке:**

```python
user_item = user_item[:1000]
```

*Количество уникальных пользователей и фильмов в выборке:*

```python
n_users = user_item.user_id.unique().shape[0]
n_items = user_item.item_id.unique().shape[0]
users = list(set(user_item['user_id']))
items = list(set(user_item['item_id']))
print ('Number of users = ' + str(n_users) + ' | Number of movies = ' + str(n_items))
```

**Составим матрицу "Пользователь-фильм":**

```python
#составим таблицу из нулей, где строки - пользователи, а столбцы - фильмы
data_matrix = np.zeros((len(users), len(items)))
data_matrix = pd.DataFrame(data_matrix, columns=items, index=users)
```

```python
#для удобства разделим колонки исходного датафрейма на два списка
users_user_item = user_item['user_id']
items_user_item = user_item['item_id']
```

```python
#заполним таблицу с нулями единицами там, где пользователь пересекается с тем фильмом, который он смотрел
for i in range(len(users_user_item)):
    row=users_user_item.loc[i]
    col=items_user_item.loc[i]
    data_matrix = data_matrix.set_value(row, col, 1)
```

**SVD**

```python
import scipy.sparse as sp
from scipy.sparse.linalg import svds
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
u, s, vt = svds(data_matrix, k = 20)
s_diag_matrix=np.diag(s)
X_pred = np.dot(np.dot(u, s_diag_matrix), vt)
```

```python
pd.options.display.float_format = '{:,.4f}'.format
pd.DataFrame(X_pred, columns=data_matrix.columns, index=users)
```

```python
print('RMSE: ' + str(sqrt(mean_squared_error(data_matrix, X_pred))))
```
